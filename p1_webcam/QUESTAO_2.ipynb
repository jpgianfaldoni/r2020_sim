```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import random
```

```python
img = cv2.imread('prova_laranjas.png')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(hsv)
plt.xticks([]), plt.yticks([]) 
plt.show()
lower = np.array([0,150,150])
upper = np.array([17,255,255])
```

```python
mask = cv2.inRange(hsv, lower, upper)
blur = cv2.GaussianBlur(hsv,(5,5),0)
morphMask = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,np.ones((30, 30)))
contornos, arvore = cv2.findContours(morphMask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
frame_out = cv2.drawContours(hsv, contornos, -1, [255, 0, 0], 3)
```

```python
cv2.imshow('Detector de circulos',frame_out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
